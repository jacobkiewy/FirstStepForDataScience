```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2, "g--")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2, "g*-")
```

```python
plt.subplot(1,2,1) # 1 sıra 2 kolon , 1. grafiği çiziyorum
plt.plot(numpyDizisi1,numpyDizisi2,"*-")
plt.subplot(1,2,2) # 1 sıra 2 kolon , 2. grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Başlığı")
figureAxes.set_ylabel("Y Ekseni Başlığı")
figureAxes.set_title("Grafik Başlığı")
```

```python
benimFigure2 = plt.figure()
figureAxes2 = benimFigure2.add_axes([0.4,0.4,0.5,0.5]) # son ikisi yükseklik ve genişlik
figureAxes2.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes2.set_xlabel("X Ekseni Başlığı")
figureAxes2.set_ylabel("Y Ekseni Başlığı")
figureAxes2.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni Küçük")
eksen2.set_ylabel("Y Ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
(benimFigure3, benimEksenler) = plt.subplots(nrows=1,ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")

plt.tight_layout()
#benimFigure
```

```python
yeniFigur = plt.figure(dpi=100)
#dpi=150

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 ** 3, label= "numpyDizisi **3 ")
yeniEksen.legend(loc=6)
```

```python
yeniFigur.savefig("BenimFigur.png",dpi=300,facecolor="g")
```
